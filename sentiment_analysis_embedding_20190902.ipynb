```python
import pandas as pd
df_health = pd.read_csv("df_health.csv", encoding="utf-8")
df_diet = pd.read_csv("df_diet.csv", encoding="utf-8")
df_slim = pd.read_csv("df_slim.csv", encoding="utf-8")
df_sport = pd.read_csv("df_sport.csv", encoding="utf-8")
df_beautiful = pd.read_csv("df_beautiful.csv", encoding="utf-8")
df_oralcare = pd.read_csv("df_oralcare.csv", encoding="utf-8")
df_foothealth = pd.read_csv("df_foothealth.csv", encoding="utf-8")
df_encyclopedia = pd.read_csv("df_encyclopedia.csv", encoding="utf-8")
```

```python
# 貼標
healthy_list = [df_health, df_diet, df_slim, df_sport, df_beautiful, df_oralcare, df_foothealth, df_encyclopedia]
k = 0
for df in healthy_list:
    df["target"] = str(k)
    k += 1
```

```python
# 合併已貼標的DataFrame
train_df = pd.concat(healthy_list, axis=0, ignore_index=True)
train_df = train_df.drop("Unnamed: 0", axis=1)
train_df
```

```python
# jieba 切割
# 將文章斷字
from jieba.analyse import extract_tags
from jieba import cut, set_dictionary, load_userdict, lcut
import os
import re


if not os.path.exists("dict.txt.big"):
    print("幫你下載大辭典")
    url = "https://github.com/fxsjy/jieba/raw/master/extra_dict/dict.txt.big"
    urlretrieve(url,"dict.txt.big")
else:
    print("你已經下載過了")

set_dictionary("dict.txt.big")

stopwords = [line.strip() for line in open("stopword.txt", 'r', encoding='utf-8').readlines()] 

def stopword(cutted_list) -> list:
    stop_list = []
    for i in cutted_list:
        if i not in stopwords:
            stop_list.append(i)
    return stop_list

def jieba_tokenize(text) -> list:
    text = re.sub("[\s+\.\!\/_,$%^*(+\"\']+|[+——！，。？、~@#￥%……&*（）【／】]+", "",text)  # 取代所有標點
    return lcut(text) 

train_df["cut_stop_list"] = train_df["content"].apply(jieba_tokenize).apply(stopword)

train_df["cut_stop_str"] = pd.Series(" ".join(i) for i in train_df["cut_stop_list"])
```

```python
# 建立Sequential 神經網路: embedding, flatten, relu, dropout, softmax
from keras.models import Sequential
from keras.layers import Embedding
from keras.layers import Dense, Flatten, Dropout
model = Sequential()
model.add(Embedding(2000, 32, input_length=100))
model.add(Flatten())
model.add(Dense(256, activation="relu"))
model.add(Dropout(0.25))
model.add(Dense(8, activation="softmax"))
model.summary()
```

```python
# 資料預處理 提取前精華的兩千字，每個字給予一個數字當代表
from keras.preprocessing.text import Tokenizer
tok = Tokenizer(num_words=2000)
tok.fit_on_texts(train_df["cut_stop_str"])
print(tok.word_index)
```

```python
# 根據 "精華兩千字" 將每篇文章序列化
train_df_seq = tok.texts_to_sequences(train_df["cut_stop_str"])
# test_df_seq = tok.texts_to_sequences(test_df["article"])
pd.DataFrame(train_df_seq)
```

```python
# 公平起見，每篇文章提取最後100個關鍵詞，截長補短(補 0)
from keras.preprocessing.sequence import pad_sequences
train_df_pad = pad_sequences(train_df_seq, maxlen=100)
# test_df_pad = pad_sequences(test_df_seq, maxlen=100)
pd.DataFrame(train_df_pad)
```

```python
# 編譯model，此處應選擇 categorical_crossentropy

model.compile(loss="categorical_crossentropy",
              optimizer="adam",
              metrics=["accuracy"])
```

```python
# 對 y 做 One-Hot encoding
from keras.utils.np_utils import to_categorical
y_train_cat = to_categorical(train_df["target"])
print(y_train_cat)
```

```python
# 開始跑 MLP模型
model.fit(train_df_pad, y_train_cat,
          batch_size=200, epochs=10,
          validation_split=0.1, verbose=2)
```
