```python
import pandas as pd
df = pd.read_table('log_group28.csv', sep=",")
df = df.drop(columns=["tag", "url"])
logs_list = df.values.tolist()
logs_list.sort(key=lambda x: (x[0], x[1]), reverse=True)
logs_list
```

```python
user_dict = {}

for line in logs_list:
    userid, logtime, tag= line[0], line[1], line[2]
#     print(userid, logtime, tag, url)
    if userid not in user_dict:
        user_dict[userid]=[]
    
    user_dict[userid].append((tag, logtime))
user_dict
        
        
```

```python
def weighting(user_dict:dict) -> dict:
    
    for key in user_dict.keys():
    # 按照點擊文章時間加權(ex: 最新3篇*1, 次新3篇*0.5, 再次三篇*0.2), 同key 相加
        wt_dict = {}
        print(len(user_dict[key]))
        for i in range(len(user_dict[key])):
            k = user_dict[key][i][0]
            
            v = (1 if i in [0,1,2] else 0.5 
                    if i in [3,4,5] else 0.2 )
            wt_dict[k] = v if k not in wt_dict else wt_dict[k]+v
            if i == 8: break
        print(f"user {key}: ",wt_dict)
weighting(user_dict)
```
